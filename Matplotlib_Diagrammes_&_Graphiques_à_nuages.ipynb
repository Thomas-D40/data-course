```python
import pandas as pd

fandango = pd.read_csv('/gdrive/MyDrive/python-data-course/fandango_score_comparison.csv')

fandango.columns

cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']

reviews_norm = fandango[cols]

# fandango.head()
```

# Diagrammes en barre

```python
import matplotlib.pyplot as plt

# plt.bar() ou Axes.bar()

fig, ax = plt.subplots()

# axes.bar(left, height)

# Position côté gauche de chaque barre
import numpy as np

bar_position = np.arange(5) + 1

# Hauteur des barres: moyenne des notes du premier film du dataset

num_cols = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']

bar_height = fandango[num_cols].iloc[0].values

ax.bar(bar_position, bar_height, color='b', width=0.5)

plt.show()
```

# Aligner l'intitulé et le trait sur l'axe x

```python
import matplotlib.pyplot as plt

# plt.bar() ou Axes.bar()

fig, ax = plt.subplots()

tick_positions = range(1,6)

num_cols = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']

bar_height = fandango[num_cols].iloc[0].values

ax.set_xticks(tick_positions)
ax.set_xlabel('Source de notations')
ax.set_ylabel('Note motenne')
ax.set_title('Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron (2015)')

ax.set_xticklabels(num_cols, rotation=90)
ax.bar(bar_position, bar_height, color='b', width=0.5)

plt.show()
```

# Diagrammes en barre horizontale

```python
# Axes.barh(bottom, width)
```

```python
fig, ax = plt.subplots()

bar_width = fandango[num_cols].iloc[0].values

bar_positions = np.arange(5) + 1

ax.barh(bar_positions, bar_width)

tcik_positions = range(1,6)

ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)

ax.set_ylabel('Source de notation')
ax.set_xlabel('Note moyenne')
ax.set_title('Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron (2015)')

plt.show()
```

# Graphique à nuage de points

```python
# Axes.scatter(x, y)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(fandango["Fandango_Ratingvalue"], fandango["RT_user_norm"])

ax.set_xlabel('Fandango')
ax.set_ylabel('Rotten Tomatoes')

plt.show()
```

# Intervertir les axes

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,10))

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.scatter(fandango["RT_user_norm"], fandango["Fandango_Ratingvalue"], c='red')
ax2.scatter(fandango["Fandango_Ratingvalue"], fandango["RT_user_norm"], c='blue')

# ax.set_xlabel('Fandango')
# ax.set_ylabel('Rotten Tomatoes')

plt.show()
```

# Comparaison de corrélations

```python
# Via axe.set_ylim ou set_xlim
# Par défaut matplotlib vérifie les conditions minimales

import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.scatter(fandango["Fandango_Ratingvalue"], fandango["RT_user_norm"], c='blue')
ax2.scatter(fandango["Fandango_Ratingvalue"], fandango["Metacritic_user_nom"], c='red')
ax3.scatter(fandango["Fandango_Ratingvalue"], fandango["IMDB_norm"], c='green')

ax1.set_xlim(0,5)
ax1.set_ylim(0,5)
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)
ax3.set_xlim(0,5)
ax3.set_ylim(0,5)

# ax.set_xlabel('Fandango')
# ax.set_ylabel('Rotten Tomatoes')

plt.show()
```
