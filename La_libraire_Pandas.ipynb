# Normaliser des colonnes: Ajuster les valeurs pour permettre la comparaison

Normalisation en divisant par la valeur max

```python
# Series.max()

max_calories = food_info["Energ_Kcal"].max()
max_calories

```

```python
calories_normalized = food_info["Energ_Kcal"]  / food_info["Energ_Kcal"].max()
calories_normalized
```

Exercice: Normaliser protein & lipid

```python
normalized_protein = food_info["Protein_(g)"] / food_info["Protein_(g)"].max()
normalized_fat = food_info["Lipid_Tot_(g)"] / food_info["Lipid_Tot_(g)"].max()

normalized_fat, normalized_protein
```

# Créer une nouvelle colonne dans le DataFrame

```python
iron_gramms = food_info["Iron_(mg)"] / 1000

food_info["Iron_(g)"] = iron_gramms

food_info[["Iron_(mg)","Iron_(g)"]][:5]
```

Exercice: Ajouter "Normalized_Protein" & "Normalized_Fat" au dataframe

```python
food_info["Normalized_Protein"] = food_info["Protein_(g)"] / food_info["Protein_(g)"].max()
food_info["Normalized_Fat"] = food_info["Lipid_Tot_(g)"] / food_info["Lipid_Tot_(g)"].max()

food_info[["Normalized_Protein","Protein_(g)","Normalized_Fat","Lipid_Tot_(g)"]]
```

Indice nutritionnel normalisé

```python
# Score = 2 * Normalized_Protein - 0.75 * Normalized_Fat
```

```python
food_info["Norm_Nutr_Index"] = 2 * food_info["Normalized_Protein"] - 0.75 * food_info["Normalized_Fat"]
food_info[["Norm_Nutr_Index", "Normalized_Protein", "Normalized_Fat"]]
```
