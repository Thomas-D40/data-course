# Index personnalisé via dataframe.set_index()

Par défaut, pandas crée un nouvel index pour le dataframe et ajoute une colonne correspondant à l'ancien index

```python
fandango_films = fandango.set_index("FILM", drop=False, inplace=False)
fandango_films.head()
```

```python
fandango_films.loc["Cinderella (2015)"]
fandango_films["Cinderella (2015)":]
```

```python
movies = ["Top Five (2014)", "Black Sea (2015)", "Taken 3 (2015)"]
fandango_films.loc[movies]
```

## Appliquer une fonction à toutes les colonnes avec apply()

```python
# apply(fonction à appliquer, axis)
```

```python
import numpy as np

types = fandango_films.dtypes
types
```

```python
# Filtrer les nombres les float64

float_columns = types[types.values == 'float64'].index


float_df = fandango_films[float_columns]
float_df

# Application de l'écart type

deviations = float_df.apply(lambda x: np.std(x))
deviations = float_df.apply(np.std)

deviations.head()

# Exercice / 2
exo_df = float_df.apply(lambda x: x/2)
exo_df.head()
```

## Apply sur les lignes

Nécessite de rajouter axis=1 pour l'appliquer sur les lignes plutôt que les colonnes

```python
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
rt_mt_user.apply(np.std, axis=1)
```

```python
# Exercice

rt_mt_means = float_df[["RT_user_norm", "Metacritic_user_nom"]].apply(np.mean, axis=1)
rt_mt_means.head()
```
