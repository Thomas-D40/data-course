```python
# Convertir l'age en valeur numérique
```

```python
data["Age"].value_counts()
```

```python
import pandas as pd

def format_age(age_str):
  if pd.isnull(age_str):
    return None
  age_str = age_str.replace("+","")
  age_to_take = age_str.split(" ")[0]
  return int(age_to_take)
```

```python
data["int_age"] = data['Age'].apply(format_age)
data["int_age"].describe()
```

```python
import pandas as pd

def format_income(income_str):
  if (pd.isnull(income_str)):
    return None
  income_str = income_str.split(" ")[0]
  if income_str == "Prefer":
    return None
  income_str = income_str.replace("$","")
  income_str = income_str.replace(",","")
  return int(income_str)

data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(format_income)
data["int_income"].describe()
```

Regarder de quelle manière les personnes gagnant moins de 150000 dollars voyages

```python
less_than_150000 = data["int_income"] < 150000
more_than_150000 = data["int_income"] > 150000

less_than_150000_travel_counts = data[less_than_150000]["How far will you travel for Thanksgiving?"].value_counts() / data[less_than_150000].shape[0] * 100
more_than_150000_travel_counts = data[more_than_150000]["How far will you travel for Thanksgiving?"].value_counts() / data[more_than_150000].shape[0] * 100

less_than_150000_travel_counts, more_than_150000_travel_counts
```
