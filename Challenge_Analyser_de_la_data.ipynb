# Nombre d'étudiants par catégorie de Major

- Retourner les valeurs uniques de Major_category via la méthode unique()
- Pour chaque valeur, retourner toutes les lignes de cette major_category
- Calcul total étudiant représentant cette major
- Garder en mémoire sous forme d'un dictionnaire

```python
categories_list = pd.unique(recent_grads["Major_category"])
```

```python
category_student_number = {}

for category in categories_list:
  major_category_df = recent_grads[recent_grads["Major_category"] == category]
  total_grads_in_category = major_category_df["Total"].sum()
  category_student_number[category] = total_grads_in_category
```

Exercice:
- Créer une fonction qui utilise la colonne Total pour calculer le nombre d'étudiant pour chaque catégorie dans chaque dataset
- Stocker le résultat dans 2 dictionnaires

- Faire l'approche par table_pivot

```python
def count_student_by_major_category(df):
  category_student_number = {}

  if "Major_category" not in df.columns:
    return category_student_number

  categories_list = pd.unique(df["Major_category"])
  for category in categories_list:
    major_category_df = df[df["Major_category"] == category]
    total_grads_category = major_category_df["Total"].sum()
    category_student_number[category] = total_grads_category

  return category_student_number
```

```python
recent_grads_student_number = count_student_by_major_category(recent_grads)
all_ages_student_number = count_student_by_major_category(all_ages)
```

# Comparer les datasets

- Via boucle, parcourir toutes les majors
- Pour chaque major dans chaque dataset, récupérer la colonne "Unemployment rate"
- Si recent_grads < all_ages, incrémenter de 1

```python
import numpy as np

recent_grads_major_list = recent_grads["Major"].unique()
all_ages_major_list = recent_grads["Major"].unique()

len(recent_grads_major_list), len(all_ages_major_list), len(np.intersect1d(recent_grads_major_list.astype(str), all_ages_major_list.astype(str)))
```

```python
recent_grads.columns
```

```python
rg_lower_count = 0

for major in recent_grads_major_list:
  recent_grads_unemployement_rate = recent_grads[recent_grads["Major"] == major]["Unemployment_rate"].values
  all_ages_unemployement_rate = all_ages[all_ages["Major"] == major]["Unemployment_rate"].values
  if recent_grads_unemployement_rate < all_ages_unemployement_rate:
    rg_lower_count += 1

rg_lower_count
```
