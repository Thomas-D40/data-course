```python
import pandas as pd

data = pd.read_csv('/gdrive/MyDrive/python-data-course/unrate.csv')

data['DATE'] = pd.to_datetime(data['DATE'])

data.head(12)
```

# Matplotib pour les graphiques linéaires

```python
import matplotlib.pyplot as plt

first_year = data[:12]

plt.plot(first_year['DATE'], first_year['UNRATE'])
plt.show()
```

# Créer une rotation sur les données de l'axe x plt.xticks()

```python
import matplotlib.pyplot as plt

first_year = data[:12]

plt.plot(first_year['DATE'], first_year['UNRATE'])
plt.xticks(rotation=90)
plt.show()
```

# Ajout des labels & titres

```python
import matplotlib.pyplot as plt

first_year = data[:12]

plt.plot(first_year['DATE'], first_year['UNRATE'])
plt.xticks(rotation=90)
plt.xlabel('Mois')
plt.ylabel('Taux de chômage (%)')
plt.title('Taux de chômage en 1948 aux USA')
plt.show()
```

Exercice:
Fgiure 10 * 6
1948 red

```python
list_colors = ['red','blue','green','orange','black']

fig = plt.figure(figsize=(10,6))

for i in range(5):
  plt.plot(data[i*12:(i+1)*12]['MONTH'],data[i*12:(i+1)*12]['UNRATE'], c=list_colors[i])

plt.show()
```

# Ajout du label via le paramètre label dans plt.plot

```python
list_colors = ['red','blue','green','orange','black']

fig = plt.figure(figsize=(10,6))

for i in range(5):
  plt.plot(data[i*12:(i+1)*12]['MONTH'],data[i*12:(i+1)*12]['UNRATE'], c=list_colors[i], label = str(1948 + i))

plt.legend(loc='best')
plt.title('Taux de chômage par mois')
plt.xlabel('Mois')
plt.ylabel('Taux de chômage (%)')
plt.show()
```
