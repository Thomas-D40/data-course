# Trouver les valeurs manquantes

None en Python vs NAN en Pandas

```python
# pandas.isnull("nom colonne") => renvoie un objet series contenant des booleans
```

```python
columns_with_missing_values = []

sex = titanic_survival["sex"]
sex_is_null = pd.isnull(sex)

sex_null_is_true = sex[sex_is_null]
type(sex_null_is_true)

for column in titanic_survival.columns:
  column_is_null = pd.isnull(titanic_survival[column])
  column_contains_null = titanic_survival[column][column_is_null]
  if len(column_contains_null) > 0:
    columns_with_missing_values.append(column)

columns_with_missing_values
```

# Appliquer une fonction à une ligne via axis=1 sur apply

```python
def is_minor(row):
  return row["age"] < 18
```

```python
minors = titanic_survival.apply(is_minor, axis=1)
minors
```

```python
def which_class(row):
  pclass = row["pclass"]

  if pd.isnull(pclass):
    return "Unknown"
  elif pclass == 1:
    return "First Class"
  elif pclass == 2:
    return "Second Class"
  else:
    return "Third Class"
```

```python
classes = titanic_survival.apply(which_class, axis=1)
classes
```

Exercice:

- Créer une fonction qui retourne "mineur" si age < 18, "adult" si > et "unknown" si NaN
- Assigner l'apply à age_labels

```python
def is_minor_or_adult(row):
  age = row["age"]

  if pd.isnull(age):
    return "unknown"

  if age < 18:
    return "minor"
  else:
    return "adult"


age_labels = titanic_survival.apply(is_minor_or_adult, axis=1)
age_labels
```

Exercice:
- ajouter la colonne "age_labels" au dataframe
- créer une table pivot pour calculer la moyenne de chance de survie en fonction de l'âge
- assigner à age_goup_surival

```python
titanic_survival["age_labels"] = age_labels
age_group_survival = titanic_survival.pivot_table(index="age_labels", values="survived")
age_group_survival
```
